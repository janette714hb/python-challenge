```python
import pandas as pd
import os
import csv
import numpy as np
```

```python
# Store filepath in a variable
schools = "schools_complete.csv"
students = "students_complete.csv"
```

```python
# Read our Data file with the pandas library
schools_df = pd.read_csv(schools)
students_df = pd.read_csv(students)
```

```python
#Check Schools.csv
schools_df.head()
```

```python
#Check Students.csv
students_df.head()
```

```python
# Merge two dataframes using an outer join
#merge_table = pd.merge(info_pd, items_pd, on="customer_id", how="outer") merge_table
district_df = pd.merge(schools_df, students_df, on="school_name", how="outer")
district_df.head(10)
```

```python
# Count how many schools in table - check size of table for missing data
district_df.count()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = district_df.rename(columns={"school_name" : "School Name", "student_name" : "Student Name", "budget" : "Budget"})
renamed_df.head()
```

```python
# storing unique value in a variable for school name
total_schools = renamed_df["School Name"].nunique()
total_schools
```

```python
# storing unique value in a variable for total students
total_students = renamed_df["Student ID"].nunique()
total_students
```

```python
# storing unique value in a variable for total budget
total_budgets = schools_df["budget"].sum()
total_budgets
```

```python
# storing average math score
avg_math = round(district_df["math_score"].mean(), 2)
avg_math
```

```python
# storing average reading score
avg_reading = round(district_df["reading_score"].mean(), 2)
avg_reading
```

```python
# count number of students with passing math score over 70% 
math_pass = district_df[district_df["math_score"] > 70].count()["school_name"]
math_pass
```

```python
# count number of students with passing reading score over 70%
read_pass = district_df[district_df["reading_score"] > 70].count()["school_name"]
read_pass
```

```python
# percentage of students who passed math with over 70% - round to 100th
mathpass_percent = round(math_pass / total_students * 100, 2)
mathpass_percent
```

```python
# percentage of students who passed reading with over 70% - round to 100th
readpass_percent = round(read_pass / total_students * 100, 2)
readpass_percent
```

```python
# Overall passing rate (average of math and reading passing rates) - round to 100th
overall_percent = round(((mathpass_percent + readpass_percent) / 2), 2)
overall_percent
```

```python
#Create data table for District results
district_results = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students],"Total Budget":[total_budgets], "Average Math Score":[avg_math], "Average Reading Score":[avg_reading], "% Passing Math":[mathpass_percent], "% Passing Reading":[readpass_percent], "Overall Passing rate":[overall_percent]})
district_results
```

```python
# Determine the school type
school_types = schools_df.set_index(["school_name"])["type"]
school_types
```

```python
# list total number of students per high school
students_per_school = district_df.groupby(["school_name"]).count()["student_name"]
students_per_school
```

```python
# list school budget per school
per_school_budget = district_df.groupby(["school_name"]).mean()["budget"]
per_school_budget
```

```python
# calculate per student budget
per_student_budget = per_school_budget / students_per_school
per_student_budget
```

```python
# calculate average math score per student at each high school
avg_math_score_per_student = district_df.groupby(["school_name"]).mean()["math_score"]
avg_math_score_per_student
```

```python
# calculate average reading score per student at each high school
avg_read_score_per_student = district_df.groupby(["school_name"]).mean()["reading_score"]
avg_read_score_per_student
```

```python
# count number of students by high school with passing math score over 70% 
math_pass_school = district_df[district_df["math_score"] > 70].groupby("school_name").count()["student_name"]
math_pass_school
```

```python
# percentage of students who passed math
math_pass_percent_school = round(math_pass_school / students_per_school * 100, 1)
math_pass_percent_school
```

```python
# count number of students by high school with passing reading score over 70% 
read_pass_school = district_df[district_df["reading_score"] > 70].groupby("school_name").count()["student_name"]
read_pass_school
```

```python
# percentage of students who passed math
read_pass_percent_school = round(read_pass_school / students_per_school * 100, 1)
read_pass_percent_school
```

```python
# overall passing rate percentage by school 
avg_total_score_per_student = (read_pass_percent_school + math_pass_percent_school) / 2
avg_total_score_per_student
```

```python
#Create data table for School results by High School
School_Summary = pd.DataFrame({"School Type":school_types, "Total Students":students_per_school,
                               "Total School Budget":per_school_budget, "Per Student Budget":per_student_budget,
                                "Average Math Score":avg_math_score_per_student, "Average Reading Score":avg_read_score_per_student,
                                 "% Passing Math":math_pass_percent_school, "% Passing Reading":read_pass_percent_school,
                                 "Overall Passing Rate":avg_total_score_per_student})
School_Summary
```

```python
#format budget as currency and math scores == file_df["other"] = file_df["other"].map("{:.2f}".format)

School_Summary["Average Reading Score"] = School_Summary["Average Reading Score"].map("{:.2f}".format)
School_Summary["Average Math Score"] = School_Summary["Average Math Score"].map("{:.2f}".format)
School_Summary["Total School Budget"] = School_Summary["Total School Budget"].map("${:,.0f}".format)
School_Summary["Per Student Budget"] = School_Summary["Per Student Budget"].map("${:,.0f}".format)

School_Summary
```

```python
# Top 5 performing schools(By passing rates)
Top5_Schools = School_Summary.sort_values(["Overall Passing Rate"], ascending = False)
Top5_Schools.head(5)
```

```python
# Bottom 5 performing schools(By passing rates)
Top5_Schools = School_Summary.sort_values(["Overall Passing Rate"], ascending = False)
Top5_Schools.tail(5)
```

```python
# Calculate average math score by grade by school == district_df
ninth_gr_score_math = district_df[district_df["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_gr_score_math = district_df[district_df["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_gr_score_math = district_df[district_df["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_gr_score_math = district_df[district_df["grade"] == "12th"].groupby("school_name").mean()["math_score"]
```

```python
# Create table for Calculate average math score by grade by school
math_score_by_grade_df = pd.DataFrame({"9th":ninth_gr_score_math, "10th":tenth_gr_score_math,
                               "11th":eleventh_gr_score_math,"12th":twelfth_gr_score_math})
math_score_by_grade_df
```

```python
#format math scores to one decimal place
math_score_by_grade_df["9th"] = math_score_by_grade_df["9th"].map("{:.1f}".format)
math_score_by_grade_df["10th"] = math_score_by_grade_df["10th"].map("{:.1f}".format)
math_score_by_grade_df["11th"] = math_score_by_grade_df["11th"].map("{:.1f}".format)
math_score_by_grade_df["12th"] = math_score_by_grade_df["12th"].map("{:.1f}".format)

math_score_by_grade_df
```

```python
# Calculate average reading score by grade by school == district_df
ninth_gr_score_read = district_df[district_df["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
tenth_gr_score_read = district_df[district_df["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
eleventh_gr_score_read = district_df[district_df["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
twelfth_gr_score_read = district_df[district_df["grade"] == "12th"].groupby("school_name").mean()["reading_score"]
```

```python
# Create table for Calculate average math score by grade by school
read_score_by_grade_df = pd.DataFrame({"9th":ninth_gr_score_read, "10th":tenth_gr_score_read,
                               "11th":eleventh_gr_score_read,"12th":twelfth_gr_score_read})
read_score_by_grade_df
```

```python
# format reading scores to one decimal place
read_score_by_grade_df["9th"] = read_score_by_grade_df["9th"].map("{:.1f}".format)
read_score_by_grade_df["10th"] = read_score_by_grade_df["10th"].map("{:.1f}".format)
read_score_by_grade_df["11th"] = read_score_by_grade_df["11th"].map("{:.1f}".format)
read_score_by_grade_df["12th"] = read_score_by_grade_df["12th"].map("{:.1f}".format)

read_score_by_grade_df
```

```python
# Scores by school spending
spend_bins = [0, 585, 615, 645, 675]
spend_ranges = ["<585", "585-615", "615-645", "645-675"]
```

```python
# Categorize spending based on spending bins
pd.cut(per_student_budget, spend_bins, labels = spend_ranges)
```

```python
# Categorize spending based on spending bins
School_Summary["Spending Ranges Per Student"] = pd.cut(per_student_budget, spend_bins, labels = spend_ranges)
School_Summary
```

```python
# create a groupby object for school performance on spending ranges
School_Summary = School_Summary.groupby("Spending Ranges Per Student")
print(School_Summary["Average Math Score"].value_counts())
```
